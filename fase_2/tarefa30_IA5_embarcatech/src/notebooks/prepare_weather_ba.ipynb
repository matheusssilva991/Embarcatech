## Consolidação dos Dados de Todos os Anos
Nesta etapa, todos os arquivos anuais das estações da Bahia são lidos, filtrados e consolidados em um único DataFrame para análise.

```python
# Leitura, filtragem e concatenação dos dados meteorológicos de todos os anos para as estações da Bahia
weather_df = pd.DataFrame()
years = [year for year in range(2000, 2026)]
columns_to_keep = [
    'DATA (YYYY-MM-DD)',
    'Hora UTC',
    'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)',
    'PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)',
    'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)',
    'UMIDADE RELATIVA DO AR, HORARIA (%)',
    'ESTACAO'
    ]

rename_dict = {
    'DATA (YYYY-MM-DD)': 'data',
    'Hora UTC': 'hora',
    'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'precipitacao',
    'PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)': 'pressao',
    'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)': 'temperatura',
    'UMIDADE RELATIVA DO AR, HORARIA (%)': 'umidade',
    'ESTACAO': 'estacao',
}

for year in years:
    weather_csv_path = f"{data_path}/weather_{year}/weather_{year}.csv"
    temp_df = pd.read_csv(weather_csv_path)
    temp_df = temp_df[columns_to_keep]
    temp_df = temp_df[temp_df['ESTACAO'].isin(id_stations_BA)]
    temp_df = temp_df.rename(columns=rename_dict)
    weather_df = pd.concat([weather_df, temp_df], ignore_index=True)

# Adicionar coluna de estado e cidade a partir do cadastro de estacoes
stations_lookup = stations_df[['id_station', 'state', 'city_station']].copy()
stations_lookup['id_station'] = stations_lookup['id_station'].astype(str)
weather_df['estacao'] = weather_df['estacao'].astype(str)
weather_df = weather_df.merge(
    stations_lookup,
    left_on='estacao',
    right_on='id_station',
    how='inner',
)
weather_df = weather_df.rename(columns={'state': 'estado', 'city_station': 'cidade'})
weather_df = weather_df.drop(columns=['id_station'])

weather_df.head()
```

```python
# Informações gerais do DataFrame consolidado
weather_df.info()
```

## Otimização e Conversão de Tipos
Os dados são otimizados para reduzir o uso de memória e a coluna de data/hora é criada para facilitar análises temporais.

```python
# Otimização dos tipos de dados e criação da coluna datetime
weather_df = weather_df.astype({
    'precipitacao': 'float32',
    'pressao': 'float32',
    'temperatura': 'float32',
    'umidade': 'float32',
    'estacao': 'str'
})

# Garantir que as colunas 'data' e 'hora' sejam string antes de concatenar
weather_df['data'] = weather_df['data'].astype(str)
weather_df['hora'] = weather_df['hora'].astype(str)
weather_df['datetime'] = pd.to_datetime(weather_df['data'] + ' ' + weather_df['hora'], errors='coerce')
weather_df = weather_df.drop(columns=['data', 'hora'])
weather_df = weather_df.drop_duplicates()
weather_df.info()
```

```python
# Visualização das primeiras linhas do DataFrame final
weather_df.head()
```

## Tratamento de Valores Faltantes
Aplicação de técnicas para lidar com valores ausentes, garantindo a qualidade dos dados para análises e modelos de IA.

```python
# Checagem de valores faltantes em cada coluna
for column in weather_df.columns:
    print(f"{column}: {weather_df[column].isnull().sum()} missing values")
```

```python
# Tratamento de valores faltantes para uso em IA
 # Preenche precipitação com zero (ausência = sem chuva)
weather_df['precipitacao'] = weather_df['precipitacao'].fillna(0)

# Preenche pressao, temperatura e umidade com média móvel de 7 períodos para preservar tendências temporais
for col in ['pressao', 'temperatura', 'umidade']:
    weather_df[col] = weather_df[col].fillna(weather_df[col].rolling(window=7, min_periods=1).mean())
    # Interpola valores faltantes restantes (linear)
    weather_df[col] = weather_df[col].interpolate(method='linear', limit_direction='both')
    # Preenche o que sobrar com a média global da coluna
    weather_df[col] = weather_df[col].fillna(weather_df[col].mean())

# Remove linhas onde datetime está faltando (essencial para séries temporais)
weather_df = weather_df.dropna(subset=['datetime'])

# Exibe informações após o tratamento
weather_df.info()
```

```python
# Checagem final de valores faltantes após o tratamento
for column in weather_df.columns:
    print(f"{column}: {weather_df[column].isnull().sum()} missing values")
```
